```python
import pandas as pd
from sqlalchemy import create_engine
engine=create_engine("postgresql://postgres:tiger@localhost:5432/dataanalyst")
c = pd.read_sql("SELECT * FROM customers", engine)
t = pd.read_sql("SELECT * FROM transactions", engine)
# customers table cleaning process using UDUSBRR
# 1. Understand the data 2. Datatype checking
# c.info()
# 3. Uniqueness checking
# c['customer_id'].duplicated().sum()
# 4.Standardization
# c['marital_status'].value_counts()
# c['prefered_order_cat'].value_counts()
c=c.replace('Mobile Phone', 'Mobile')
# 5) Blank value checking
c['tenure'] = c['tenure'].fillna(c['tenure'].median(),)
# c.isnull().sum()
# 6) Business rule checking
# c['tenure'].unique()
# c['number_of_device_registered'].unique()
# c['number_of_address'].unique()
# c['churn'].value_counts()
# 7) Remove duplicates
# c.duplicated().sum()
c.drop_duplicates(inplace=True)



# transactions table cleaning process using UDUSBRR
# 1. Understand the data 2. Datatype checking
# t.info()
# 3. Uniqueness checking
# t['customer_id'].duplicated().sum()
# 4.Standardization
# 5) Blank value checking
t['warehouse_to_home'] = t['warehouse_to_home'].fillna(t['warehouse_to_home'].median())
t['day_since_last_order'] = t['day_since_last_order'].fillna(t['day_since_last_order'].median())
# t.isnull().sum()
# 6) Business rule checking
# t['warehouse_to_home'].unique()
# t['satisfaction_score'].unique()
# t['complain'].unique()
# t['day_since_last_order'].unique()
# t['cashback_amount'].value_counts()
# t[t['cashback_amount']<0]

# pd.set_option('display.max_rows', None)
# pd.reset_option('display.max_rows')

# merge=pd.merge(c,t, on='customer_id', how='inner')
# merge.loc[
#     merge['day_since_last_order'] > (merge['tenure'] * 30),
#     ['tenure', 'day_since_last_order','customer_id']
# ]
# 7) Removing duplicates

# t.duplicated().sum()
t.drop_duplicates(inplace=True)

# EDA
merged=pd.merge(c,t, on='customer_id', how='inner')
merged.shape
merged.groupby('churn')['tenure'].mean()
merged.groupby('marital_status')['churn'].mean()*100
merged.groupby('number_of_device_registered')['churn'].mean()*100
merged.groupby('number_of_address')['churn'].mean()*100
merged=merged.replace('Mobile Phone','Mobile',regex=True)
merged['prefered_order_cat'].value_counts()
merged.groupby('prefered_order_cat')['churn'].mean()*100
merged.groupby('warehouse_to_home')['churn'].mean()*100
merged.groupby('satisfaction_score')['churn'].mean()*100
merged.groupby('complain')['churn'].mean()*100
merged.groupby('churn')['day_since_last_order'].mean()
merged.groupby('churn')['cashback_amount'].mean()

c.to_csv('customers_clean.csv', index=False)
t.to_csv('transactions_clean.csv', index=False)

c.to_sql('customers_clean',engine,if_exists='replace', index=False)
t.to_sql('transactions_clean',engine,if_exists='replace',index=False)

c.info()
```
